```python
from sympy import *
init_printing()
```

```python
x=Symbol('x')
```

```python
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
f
```

```python
# попытка найти корни уравнения
solveset(Eq(-12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30, 0),x)

# получился интересный результат, но не уверен, что он правильный
```

```python
# построение графика функции по умолчанию в пределах от -10 до 10
g=plot(f,x, title='График функции в пределах х(-10;10)')
```

```python
# построение графика функции в пределах от -100 до 100 для большей нагядности
g=plot(f,(x,-100,100),title='График функции в пределах х(-10;10)', line_color='red')
```
